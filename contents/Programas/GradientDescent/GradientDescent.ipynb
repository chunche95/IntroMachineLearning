```python
import numpy as np
import scipy as sc 
import matplotlib.pyplot as plt 
```

```python
func = lambda th: np.sin(1/2*th[0]**2-1/4*th[1]**2+3)* np.cos(2*th[0]+1 - np.e**th[1])
res=100
_X = np.linspace(-2, 2, res)
_Y = np.linspace(-2, 2, res)
_Z = np.zeros((res, res))

for ix, x in enumerate(_X):
    for iy, y in enumerate (_Y):
        _Z[iy, ix] = func([x,y])

# plt.contour?
# Maya sólida
plt.contourf(_X,_Y,_Z, 100)
plt.colorbar()
# plt.show()

Theta = np.random.rand(2)*4-2
_T = np.copy(Theta)
h = 0.01
#ratio de aprendizaje 
lr = 0.001

plt.plot(Theta[0], Theta[1], "o", c="white")

grad = np.zeros(2)
# Usamos derivadas parciales para ir descendiendo la pendiente
for _ in range(10000):
    for it, th in enumerate(Theta):
        _T = np.copy(Theta)
        _T[it] = _T[it]+h
        # Derivada de la pendiente del eje Z0
        deriv = (func(_T) - func(Theta))/h
        grad[it] = deriv
    # Vector que actualiza los parámetros y minimizar la función
    Theta = Theta - lr * grad 
    # print(func(Theta))
    if(_ % 50 ==0):
        plt.plot(Theta[0], Theta[1], '.', c="orange")
plt.plot(Theta[0], Theta[1], "o", c="blue")
plt.show()
```
