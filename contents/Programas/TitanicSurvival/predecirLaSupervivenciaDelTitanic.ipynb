```python
# Librerías usadas
import numpy as np 
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Importando la DATA
# Importamos los datos desde la página de Kaggle, guardamos una copia en nuestro equipo y a su vez almacenamos los datos en variables Pandas para poderlos manipular a lo largo del programa.
url_test =  'D:/Descargas/IntroMachineLearning/contents/Programas/Titanic_Data/test.csv'
url_trian = 'D:/Descargas/IntroMachineLearning/contents/Programas/Titanic_Data/train.csv'

df_test = pd.read_csv(url_test)
df_train = pd.read_csv(url_trian)

# Se guardan los datos en un archivo para siempre tenerlos disponibles
dirTest = 'D:/Descargas/IntroMachineLearning/contents/Programas/Titanic_Data/Titanic_test(Copia).csv'
dirTrain = 'D:/Descargas/IntroMachineLearning/contents/Programas/Titanic_Data/Titanic_train(Copia).csv'

df_test.to_csv(dirTest)
df_train.to_csv(dirTrain)

# Importamos los datos de los archivos CSV almacenados
```

```python
# Entendiendo los Data
# Verificamos la cantidad de datos con la que contamos, el tipo de datos, la cantidad faltante que tiene el dataset y por último las estadísticas del mismo. 

# Cantidad de datos del DataSet
print("-----------------------")
print('Cantidad de datos:')
print("-----------------------")
print(df_test.shape)
print(df_train.shape)

# Verifico el tipo de datos contenido en ambos dataset.
print("-----------------------")
print('Tipos de datos:')
print("-----------------------")
print(df_test.info())
print(df_train.info())

# Verifico los datos faltantes de los dataset.
print("-----------------------")
print('Datos faltantes:')
print("-----------------------")
print(pd.isnull(df_test).sum())
print(pd.isnull(df_train).sum())

# Verifico las estadísticas del dataset.
print("-----------------------")
print('Estadísticas del dataset:')
print("-----------------------")
print(df_test.describe())
print(df_train.describe())
```

```python
# Procesamiento de datos.
# Cambio los datos de sexos en número.
df_train['Sex'].replace(['female', 'male'],[0,1], inplace=True)
df_test['Sex'].replace(['female', 'male'],[0,1], inplace=True)

# Cambio los datos de embarque en números.
df_train['Embarked'].replace(['Q','S','C'], [0,1,2],inplace=True)
df_test['Embarked'].replace(['Q','S','C'], [0,1,2],inplace=True)

# Reemplazo los datos faltantes en la edad por la media de esta columna.
print(df_train["Age"].mean())
print(df_test["Age"].mean())

promedio = 30
df_train['Age'] = df_train['Age'].replace(np.nan, promedio)
df_test['Age'] = df_test['Age'].replace(np.nan, promedio)

# Genero grupos de acuedo a la edad: (0-8)(9-15))(16-18)(19-25)(26-40)(41-60)(61-100)
bins=[0,8,15,18,25,40,60,100]

# Creando 7 grupos
names=['1','2','3','4','5','6','7']

# Creo la matriz con los datos
df_train['Age'] = pd.cut(df_train['Age'],bins,labels = names)
df_test['Age'] = pd.cut(df_train['Age'],bins,labels = names)

# Elimino las columnas que no tienen muchos datos o faltan demasiados. 
df_train = df_train.drop(['Cabin'], axis=1, inplace=True)
df_test = df_test.drop(['Cabin'], axis=1, inplace=True)
df_train = df_train.drop(['PassangerId', 'Name', 'Ticket'], axis=1)
df_test = df_test.drop(['Name', 'Ticket'], axis=1)

# Elimino las filas con los datos perdidos.
df_train.dropna(axis=0, how='any', inplace=True)
df_test.dropna(axis=0, how='any', inplace=True)

# Verifico los datos resultantes.
print(pd.isnull(df_train).sum())
print(pd.isnull(df_test).sum())

print(df_train.shape)
print(df_test.shape)

print(df_test.head())
print(df_train.head())
```

```python
# Implementamos el logaritmo de ML, en este caso es Regresión Logística, vectores de soporte y vecinos mñas cercanos.
# Separo la columna con la información de los sobrevivientes
X = np.array(df_train.drop(['Survived'], 1))
y = np.array(df_train['Survived'])

# Separo los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Regresión logística
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
Y_pred = logreg.predict(X_test)
print('Precisión Regresión Logística:')
print(logreg.score(X_train, y_train))

# Support Vector Machines
svc = SVC()
svc.fit(X_train, y_train)
Y_pred = svc.predict(X_test)
print('Precisión Soporte de Vectores:')
print(svc.score(X_train, y_train))

# K neighbors
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
Y_pred = knn.predict(X_test)
print('Precisión Vecinos más Cercanos:')
print(knn.score(X_train, y_train))
```
